```python
neows_url = f"https://api.nasa.gov/neo/rest/v1/neo/browse?api_key={api_key}"
neows_response = requests.get(neows_url).json()
print(neows_response)
```

```python
asteroids = neows_response['near_earth_objects']  # Access the NEO data
```

```python
print(type(asteroids))
print(asteroids)
```

```python
data = []  # Initialize a list for the DataFrame

# Loop through the list of asteroids
for asteroid in asteroids:
    # Extract relevant details and handle missing fields gracefully
    data.append({
        'Asteroid ID': asteroid['id'],
        'Asteroid Name': asteroid['name'],
        'Minimal Estimated Diameter (km)': asteroid['estimated_diameter']['kilometers']['estimated_diameter_min'],
        'Absolute Magnitude': asteroid['absolute_magnitude_h'],
        'Relative Velocity (km/s)': (
            asteroid['close_approach_data'][0]['relative_velocity']['kilometers_per_second'] 
            if asteroid['close_approach_data'] else None
        ),
    })

# Convert the list of dictionaries to a DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Display the DataFrame
print(df.head())
```

```python
# Ensure clean column names and types
df = df.rename(columns=lambda x: x.strip().replace(" ", "_").lower())
```

```python
# Export DataFrame to CSV
csv_file_path = 'asteroids_data.csv'
df.to_csv(csv_file_path, index=False)
print(f"Data exported to {csv_file_path}")
```
